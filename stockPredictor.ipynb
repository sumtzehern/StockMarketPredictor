```python
# Import the yahooFinance API, pandas and os

import yfinance as yf
import pandas as pd
import os
```

```python
# More for individual stock

del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
# set target to check if the tomorrow price will be greater than ystd closing price 

sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
# Having historical is good, but too old data might not be too useful for predicting( marketshift etc)

# we clean data to show only after 1990
sp500 = sp500.loc["1990-01-01":].copy()
```

```python
# RandomForestClassifier Reduces overfitting by averaging multiple decision trees
# Pick up non-linear relationship
# If you can find linear you gonna be rich!

from sklearn.ensemble import RandomForestClassifier 

# n_estimators: might be good with higher value
# min_samples_split: protect from overfitting, higher less accurate but also less overfitt
# random_state: we would run the same random data

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"] # if we use Tomorrow,Target we are will know the future which is no going to happen in real world
model.fit(train[predictors], train["Target"])

              
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
# Turn numpy array into panda series

preds = pd.Series(preds, index=test.index)
```

```python
# Building a Backtesting System

'''
1. Fitting the model using training predictor
2. Generate prediction
3. Combine model into series
4. Combining all 
'''

def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
# Trading around 250 per year

def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    
    return pd.concat(all_predictions)
        
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
'''
2 days, last trading week, 3 month, last year, 4 years
'''

horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_column, trend_column]
```

```python
sp500 = sp500.dropna() # drop the missing column
```

```python
model = RandomForestClassifier(n_estimators=1000, min_samples_split=500, random_state=1)
```

```python
'''
1. Fitting the model using training predictor
2. Generate prediction
3. Combine model into series
4. Combining all 
'''

def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    # Custom treshold
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    
    return combined
```

```python
from sklearn.model_selection import train_test_split
from tabulate import tabulate

def train_and_backtest(data, predictors, target_column, n_estimators=1000, min_samples_split=500, random_state=1, result_csv='results.csv'):
    # Split the data into features (predictors) and target
    X = data[predictors]
    y = data[target_column]
    
    # Split the data into training and testing sets
    train_data, test_data, train_labels, test_labels = train_test_split(X, y, test_size=0.2, random_state=random_state)
    
    # Initialize the RandomForestClassifier
    model = RandomForestClassifier(n_estimators=n_estimators, min_samples_split=min_samples_split, random_state=random_state)

    # Record start time
    start_time = time.time()

    # Train the model
    model.fit(train_data, train_labels)

    # Record end time
    end_time = time.time()

    # Calculate training time
    training_time = end_time - start_time

    # Run backtest
    predictions = model.predict(test_data)

    # Calculate precision score
    precision = precision_score(test_labels, predictions)

    # Calculate distribution of predictions (for price increase and decrease)
    prediction_distribution = (pd.Series(predictions).value_counts() / len(predictions)).to_dict()

    # Create a DataFrame with the results
    result_df = pd.DataFrame({
        'n_estimators': [n_estimators],
        'min_samples_split': [min_samples_split],
        'training_time': [training_time],
        'precision': [precision],
        'increase_ratio': [prediction_distribution.get(1, 0)],
        'decrease_ratio': [prediction_distribution.get(0, 0)]
    })
    
    # Append the results to the CSV file or create a new file if it doesn't exist
    try:
        existing_results = pd.read_csv(result_csv)
        results = pd.concat([existing_results, result_df], ignore_index=True)
    except FileNotFoundError:
        results = result_df

    # Export the results to the CSV file
    results.to_csv(result_csv, index=False)
   
   # Format the result DataFrame as a table
    table = tabulate(result, headers='keys', tablefmt='pipe', showindex=False)

    # Save the formatted table to a text file
    with open('result_table.txt', 'w') as file:
        file.write(table)


    return result_df

# Example usage:
# Replace 'your_data' with your actual dataset and adjust 'predictors' and 'target_column' accordingly
result = train_and_backtest(sp500, predictors=['Close', 'Volume', 'Open', 'High', 'Low'], target_column='Target', n_estimators=1500, min_samples_split=500)
```
